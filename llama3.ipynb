```python
import transformers
import torch
```

```python
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.float16,
        "quantization_config": {"load_in_4bit": True},
        "low_cpu_mem_usage": True,
    },
)
```

```python
messages = [
    {"role": "system", "content": "You are a data scinetist who always give precise and correct answer!"},
    {"role": "user", "content": "Write a 100-word article on 'Benefits of Open-Source in AI research"},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)
```

```python
messages = [
    {"role": "system", "content": "You are a data scinetist who always give precise and correct answer!"},
    {"role": "user", "content": "Who are you?"},
    {"role": "user", "content": "Can you explain me about quantum computing in 2 lines?"},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)
```

```python
terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
```

```python




outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
print(outputs[0]["generated_text"][len(prompt):])
```
